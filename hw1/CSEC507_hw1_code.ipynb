## English Letter Frequencies

Below you can see the letter frequency and counts of English language. To calculate letter frequency and counts of **Siddhartha** Book by Hermann Hesse is used.


```python
import pandas as pd
alphabet_en = ['a', 'b', 'c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
def get_percentages(file_name, alphabet=alphabet_en):
    book = open(f'./{file_name}')
    data = book.read()
    data = data.lower()
    book.close()
    count={}
    count_sum = 0
    for char in alphabet:
        count[char] = data.count(char)
        count_sum += count[char]
    percentages = {}
    all_info = {}
    for char in alphabet:
        percentages[char] = count[char]/count_sum
        all_info[char] = [count[char], count[char]/count_sum]
    df = pd.DataFrame.from_dict(all_info, orient='index', columns=['counts', 'ratios'])
    return percentages, data, df

percentages_plain, _, df_plain = get_percentages('siddharthax8983.txt')
df_plain.sort_values(by=['ratios'], ascending=False)
```

Below you can see the ciphertext letter frequencies and counts.



```python
percentages_cipher, cipher_data, df_cipher = get_percentages('engciphertext.txt')
df_cipher.sort_values(by=['ratios'], ascending=False)
```

## Cryptanalysis of English Text
Now we obtain the reverse secret key using the frequencies from plain text.

```python
secret_key = {}
reverse_secret_key = {}
for i in range(len(alphabet_en)):
    max_value_plain = max(percentages_plain, key=percentages_plain.get)
    del percentages_plain[max_value_plain]
    max_value_cipher = max(percentages_cipher, key=percentages_cipher.get)
    del percentages_cipher[max_value_cipher]
    secret_key[max_value_plain] = max_value_cipher
    reverse_secret_key[max_value_cipher] = max_value_plain
reverse_secret_key
```

```python
pd.DataFrame.from_dict(reverse_secret_key, orient='index' ,columns=['Key'])
```

We did not get correct plaintext from ciphertext. Therefore we need to change secret key.

```python
def decipher(char):
    if char in reverse_secret_key.keys():
        return reverse_secret_key[char]
    else:
        return char
data_plain_try = map(lambda char: decipher(char), cipher_data)
res = ''.join(list(data_plain_try))
res
```
